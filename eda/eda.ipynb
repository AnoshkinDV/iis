```python
# Разделение на бинарные и непрерывные признаки
binary_features = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi']
continuous_features = ['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep', 'mobile_wt', 
                      'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']
```

```python
# Исправляем настройки matplotlib
plt.ioff()  # Отключаем интерактивный режим
plt.style.use('default')

# Создаем визуализацию (основной график)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

# Цвета для значений 0 и 1
colors = ['#FF6B6B', '#4ECDC4']  # красный и бирюзовый

for i, feature in enumerate(binary_features):
    ax = axes[i]
    
    # Получаем распределение
    distribution = df[feature].value_counts().sort_index()
    
    # Создаем барный график
    bars = ax.bar([0, 1], [distribution[0], distribution[1]], 
                    color=colors, alpha=0.7, edgecolor='black', linewidth=1)
    
    # Настройка графика
    ax.set_title(f'{feature.replace("_", " ").title()}', 
                fontsize=12, fontweight='bold')
    ax.set_xlabel('Значение')
    ax.set_ylabel('Количество')
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['Нет (0)', 'Есть (1)'])
    ax.grid(axis='y', alpha=0.3)
    
    # Добавляем значения на столбцы
    for j, (bar, value) in enumerate(zip(bars, [distribution[0], distribution[1]])):
        ax.text(j, value + 20, str(value), 
                ha='center', va='bottom', fontweight='bold')
    
    # Добавляем проценты
    pct_0 = distribution[0] / len(df) * 100
    pct_1 = distribution[1] / len(df) * 100
    ax.text(0.02, 0.98, f'{pct_0:.1f}% | {pct_1:.1f}%', 
            transform=ax.transAxes, fontsize=9,
            bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8),
            verticalalignment='top')

plt.suptitle('Распределение бинарных признаков', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('binary_features_distribution.png', dpi=300)
plt.close()  # Закрываем фигуру
    
```

```python
outliers_info = {}
for feature in continuous_features:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    outliers_count = len(outliers)
    outliers_percentage = (outliers_count / len(df)) * 100
    outliers_info[feature] = outliers_count
    print(f"{feature}: {outliers_count} выбросов из {len(df)} ({outliers_percentage:.1f}%)")
```

Проверка корелляции между признаками

```python
corr_matrix = df.corr()
```

Высокие корелляции между признаками

```python
high_corr = []
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > 0.5:
            high_corr.append((corr_matrix.columns[i], corr_matrix.columns[j], corr_matrix.iloc[i, j]))
```

```python
if high_corr:
    for feature1, feature2, corr_value in high_corr:
        print(f"{feature1} - {feature2}: {corr_value:.3f}")
else:
    print("Высокие корреляции не обнаружены")
```

График распределения целевой переменной

```python
plt.figure(figsize=(10, 6))
target_counts = df[target].value_counts().sort_index()
bars = plt.bar(target_counts.index, target_counts.values, color='skyblue', edgecolor='black', alpha=0.8)
plt.title('Распределение целевой переменной (price_range)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Ценовой сегмент', fontsize=12)
plt.ylabel('Количество телефонов', fontsize=12)
plt.xticks(target_counts.index, [f'Класс {i}' for i in target_counts.index])
plt.grid(axis='y', alpha=0.3)
```

```python
# Добавляем значения на столбцы
for bar, value in zip(bars, target_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10, 
             str(value), ha='center', va='bottom', fontsize=11, fontweight='bold')
```

```python
# порядок категорий по возрастанию индекса классов
order = df[target].value_counts().sort_index().index  # чтобы классы шли 0,1,2,... [web:26]

plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, x=target, order=order, color='skyblue', edgecolor='black', alpha=0.8)  # countplot считает частоты сам [web:21]

ax.set_title('Распределение целевой переменной (price_range)', fontsize=16, fontweight='bold', pad=20)  # заголовок [web:21]
ax.set_xlabel('Ценовой сегмент', fontsize=12)  # подпись оси X [web:21]
ax.set_ylabel('Количество телефонов', fontsize=12)  # подпись оси Y [web:21]
ax.set_xticklabels([f'Класс {i}' for i in order])  # человекочитаемые тики [web:21]

ax.grid(axis='y', alpha=0.3)  # сетка по Y [web:21]
for c in ax.containers:
    ax.bar_label(c, fmt='%d', padding=3, fontsize=11, fontweight='bold')  
plt.savefig('./eda/1_target_distribution.png', dpi=300) 
plt.show()
plt.close()
```
